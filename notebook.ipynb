```python
import numpy as np
import matplotlib.pyplot as plt
from archery import aphysix
# Constants
g = 9.81  # acceleration due to gravity, m/s^2
v0 = 87.03  # initial velocity, m/s
theta = np.pi / 6  # launch angle, radians
c = 0.002  # drag coefficient

# Define the x range
x = np.linspace(0, 400, 1000)

# Calculate y for each x
y = x * np.tan(theta) - (g * x**2) / (2 * v0**2 * np.cos(theta)**2) * ((-1 - 2*c*x + np.exp(2*c*x)) / (0.5 * (2*c*x)**2))

for idx in y:
    if idx < 0:
        idx = list(y).index(idx)
        break
    
engine = aphysix(0, 0.1)
x, y = engine.get_data(theta, v0)


# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x[:idx], y[:idx], label='Projectile motion with drag')
plt.title('Projectile Motion with Air Resistance')
plt.xlabel('x (meters)')
plt.ylabel('y (meters)')
plt.grid(True)
plt.legend()
plt.show()
```
